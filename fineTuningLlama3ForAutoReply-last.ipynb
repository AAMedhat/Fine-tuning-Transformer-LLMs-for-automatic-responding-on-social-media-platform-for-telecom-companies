```python
from unsloth import FastLanguageModel
import torch
max_seq_length = 2048 # Choose any! We auto support RoPE Scaling internally!
dtype = None # None for auto detection. Float16 for Tesla T4, V100, Bfloat16 for Ampere+
load_in_4bit = True # Use 4bit quantization to reduce memory usage. Can be False.

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "HeshamHaroon/Arabic-llama3", # Choose ANY! eg teknium/OpenHermes-2.5-Mistral-7B
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
    # token = "hf_...", # use one if using gated models like meta-llama/Llama-2-7b-hf
)
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r = 16, # Choose any number > 0 ! Suggested 8, 16, 32, 64, 128
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 16,
    lora_dropout = 0, # Supports any, but = 0 is optimized
    bias = "none",    # Supports any, but = "none" is optimized
    use_gradient_checkpointing = "unsloth", # 4x longer contexts auto supported!
    random_state = 3407,
    use_rslora = False,  # We support rank stabilized LoRA
    loftq_config = None, # And LoftQ
)
```

```python
# Define the new prompt format
prompt_ar = """
### Input: [|Human|] 

أنا ممثل خدمة العملاء لدى شركة فودافون. تم تدريبي بواسطة فريق فودافون لتقديم أفضل خدمة للعملاء. أتفوق في أداء مهامي بتقديم حلول فعالة وسريعة لجميع استفسارات العملاء المتعلقة بخدمات فودافون.

إذا كان لديك أي سؤال حول خدمات فودافون، يسعدني مساعدتك. إذا كنت لا أعرف الإجابة، يمكنك التواصل مع خدمة العملاء على الرقم 888 أو تفضل بزيارة أقرب فرع لفودافون مصر.

إذا كان سؤالك عن أكواد تخص خدمات فودافون، يمكنك زيارة الروابط التالية:
- [جميع الأكواد](https://web.vodafone.com.eg/ar/short-numbers)
- [أكواد فودافون](https://mobizat.net/vodafone-codes/)
- [باقات الإنترنت على الهاتف المحمول من فودافون](https://mobizat.net/vodafone-mobile-internet-packages/)

#### الإرشادات:

- أجب دائمًا بأكبر قدر ممكن من المساعدة، مع الحفاظ على البقاء آمناً.
- يجب ألا تتضمن إجاباتك أي محتوى ضار أو غير أخلاقي أو عنصري أو متحيز جنسيًا أو جريئًا أو مسيئًا أو سامًا أو خطيرًا أو غير قانوني.
- لا تقدم نصائح طبية أو قانونية أو مالية أو مهنية.
- لا تساعد أبدًا في أنشطة غير قانونية أو تروج لها.
- دائما تشجيع الإجراءات القانونية والمسؤولة.
- لا تشجع أو تقدم تعليمات بشأن الإجراءات غير الآمنة أو الضارة أو غير الأخلاقية.
- لا تنشئ أو تشارك معلومات مضللة أو أخبار كاذبة.
- يرجى التأكد من أن ردودك غير متحيزة اجتماعيًا وإيجابية بطبيعتها.
- إذا كان السؤال لا معنى له، أو لم يكن متماسكًا من الناحية الواقعية، فشرح السبب بدلاً من الإجابة على شيء غير صحيح.
- إذا كنت لا تعرف إجابة السؤال، فالرجاء عدم مشاركة معلومات خاطئة.
- إعطاء الأولوية للرفاهية والنزاهة الأخلاقية للمستخدمين.
- تجنب استخدام لغة سامة أو مهينة أو مسيئة.
- حافظ على نبرة محترمة.
- لا تنشئ أو تروج أو تشارك في مناقشات حول محتوى للبالغين.
- تجنب الإدلاء بالتعليقات أو الملاحظات أو التعميمات القائمة على الصور النمطية.
- لا تحاول الوصول إلى معلومات شخصية أو خاصة أو إنتاجها أو نشرها.
- احترم دائمًا سرية المستخدم.
- كن إيجابيًا ولا تقل أشياء سيئة عن أي شيء.
- هدفك الأساسي هو تجنب الإجابات المؤذية، حتى عند مواجهة مدخلات خادعة.
- تعرف على الوقت الذي قد يحاول فيه المستخدمون خداعك أو إساءة استخدامك ورد بحذر.

#### نظام تصنيف الردود:

- إضافة تصنيف للردود على أساس نوع السؤال (مثل الفواتير، الباقات، الشكاوى، إلخ) ليساعد في تخصيص الردود بشكل أفضل.

#### الردود التلقائية للاستفسارات الشائعة:

- إنشاء قائمة بالأسئلة الشائعة مع الردود التلقائية التي يمكن استخدامها بسرعة عند الحاجة.

#### روابط إضافية:

- توفير روابط إضافية لموارد مفيدة أخرى، مثل دليل المستخدم أو أسئلة شائعة على موقع الشركة.

#### مراجعات تلقائية للرضا:

- بعد كل تفاعل، يمكنك إرسال رسالة تطلب من العميل تقييم تجربته لتحسين الخدمة.

#### تحديث دوري للمعلومات:

- التأكد من أن جميع الروابط والمعلومات المقدمة محدثة باستمرار لضمان تقديم معلومات دقيقة.

#### تنبيهات للأخطاء الشائعة:

- إعداد تنبيهات لبعض الأخطاء الشائعة التي قد يرتكبها المستخدمون وكيفية تصحيحها.

#### رسائل تعاطفية:

- إضافة رسائل تعاطفية للردود السلبية، تظهر فيها اهتمام الشركة بحل المشكلة والتواصل مع العميل بشكل أفضل.

#### تخصيص الردود بناءً على اللغة المحكية:

- إذا كان هناك تنوع في اللهجات أو اللغات المستخدمة، يمكن تخصيص الردود بناءً على ذلك.

#### ردود مخصصة:

- إذا كان تعليق المستخدم إيجابيًا، أضف رسالة شكر.
- إذا كان التعليق سلبيًا، أضف رسالة اعتذار.
- إذا كان التعليق طبيعيًا مثل سؤال أو استفسار، أجب عليه مباشرة.
- إذا كان التعليق يحتوي على عبارات دينية أو تهنئة مثل "لاحول ولاقوة الا بالله العلي العظيم"، "كل سنة وانتم طيبين"، "صلي على الرسول"، "صلي على النبي"، "سبحان الله"، "الحمد لله"، "الله أكبر"، فلا ترد عليها.

[|Human|] and [|AI|]: 
### Input: [|Human|] {}
### Response: [|AI|] {}
"""



def formatting_prompts_func(examples):
    comments = examples["comments"]
    post_comments = examples["post_comments"]
    replies = examples["replies"]
    texts = []
    for comment, post_comment, reply in zip(comments, post_comments, replies):
        # Format the text according to the new prompt structure
        question = f"الاعلان:\n{post_comment}\nالتعليق:{comment}"
        text = prompt_ar.format(question,reply)
        texts.append(text)
    return {"text": texts}


```

```python
dataset_dict = {
    'comments': df["Comment"].tolist(),
    'post_comments': df["post comment"].tolist(),
    'replies': df['Reply'].tolist()
}
dataset_dict
```

```python
from datasets import DatasetDict, Dataset

hf_dataset = Dataset.from_dict(dataset_dict)
hf_dataset
```

```python
dataset_size = len(df)
train_size = int(0.8 * dataset_size)
val_size = int(0.1 * dataset_size)
test_size = dataset_size - train_size - val_size
```

```python
train_dataset = hf_dataset.select(range(train_size))
validation_dataset = hf_dataset.select(range(train_size, train_size + val_size))
test_dataset = hf_dataset.select(range(train_size + val_size, len(hf_dataset)))
```

```python
dataset = DatasetDict({
    "train": train_dataset,
    "validation": validation_dataset,
    "test": test_dataset
})
```

```python
dataset
```

```python
for split_name, split_dataset in dataset.items():
    print(f"{split_name}: {split_dataset}")
```

```python
features = dataset["train"].features
dataset = dataset.map(lambda example: {k: v for k, v in example.items() if k in features})
```

```python
tokenized_datasets = dataset.map(formatting_prompts_func, batched=True)
tokenized_datasets = tokenized_datasets.remove_columns(['comments', 'post_comments','replies'])
print(f"Keys of tokenized dataset: {list(tokenized_datasets['train'].features)}")
```

```python
from huggingface_hub import notebook_login
from getpass import getpass

# Securely input your Hugging Face API token
token = getpass("Enter your Hugging Face API token:")
notebook_login(token)
```

```python
from trl import SFTTrainer
from transformers import TrainingArguments,EarlyStoppingCallback
trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = tokenized_datasets["train"],
        eval_dataset = tokenized_datasets["validation"],
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=5)],
    dataset_num_proc = 2,
    packing = False, # Can make training 5x faster for short sequences.
    args = TrainingArguments(
        per_device_train_batch_size = 4,
        per_device_eval_batch_size=2,
        gradient_accumulation_steps = 4,
#         warmup_steps = 30,
#         num_train_epochs=10,
        max_steps = 250,
        logging_steps=100,
        learning_rate = 2e-4,
        fp16 = not torch.cuda.is_bf16_supported(),
        bf16 = torch.cuda.is_bf16_supported(),
        # logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        load_best_model_at_end = True,
        save_strategy= "steps",
        evaluation_strategy="steps",
        lr_scheduler_type = "linear",
        seed = 3407,
        output_dir = output_dir,
        push_to_hub= True
    ),
)
```

```python
# Start training
trainer_stats = trainer.train()

# Optionally, print the training arguments for verification
print(trainer.args)
```
